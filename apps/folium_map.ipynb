```python
import geopandas
```

```python
import folium
```

```python
pc4_data = geopandas.read_file('./usa_pc_len4.geojson')
pc2_data = geopandas.read_file('./world_pc_len2.geojson')
type(pc4_data)
```

```python
pc4_data
```

```python
pc2_data.head(2)
```

```python
webmap = folium.Map(location = [40, -97], zoom_start=5)
```

```python
folium.GeoJson('./usa_pc_len4.geojson', name='usa_pc4', style_function=lambda x: {'fillColor': 'orange'}).add_to(webmap)
folium.GeoJson('./world_pc_len2.geojson', name='world_pc2', style_function=lambda x: {'fillColor': 'green'}).add_to(webmap)
```

```python
for _, r in pc2_data.iterrows():
    pc2_json = geopandas.GeoSeries(r['geometry']).to_json()
    folium_data = folium.GeoJson(data=pc2_json,
                           style_function=lambda x: {'fillColor': 'orange'})
    folium.Popup(r['trimmed_pluscodes']).add_to(folium_data)
    folium_data.add_to(webmap)
```

```python
for _, r in pc4_data.iterrows():
    pc4_json = geopandas.GeoSeries(r['geometry']).to_json()
    folium_data = folium.GeoJson(data=pc4_json,
                           style_function=lambda x: {'fillColor': 'orange'})
    folium.Popup(r['trimmed_pluscodes']).add_to(folium_data)
    folium_data.add_to(webmap)
```
